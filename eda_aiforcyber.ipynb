```python
# EDA on Ransomware Files CSV
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = 'data_file.csv'  # Update this with your file path
data = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset
print("First few rows of the dataset:")
display(data.head())
```

```python
# Display basic information about the dataset
print("\nDataset Information:")
data.info()
```

```python
# Display summary statistics
print("\nSummary Statistics:")
display(data.describe())
```

```python
# Check the distribution of benign vs malicious files
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x='Benign')
plt.title('Distribution of Benign vs Malicious Files')
plt.xlabel('Benign')
plt.ylabel('Count')
plt.grid()
plt.show()
```

```python
# 2. Correlation Matrix
plt.figure(figsize=(15, 10))
numeric_cols = data.select_dtypes(include=[np.number]).columns
correlation = data[numeric_cols].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Calculate the CDF
sorted_debug_size = np.sort(data['DebugSize'])
cdf_y = np.arange(1, len(sorted_debug_size) + 1) / len(sorted_debug_size)

# Plot the CDF
plt.plot(sorted_debug_size, cdf_y, marker='.', linestyle='none', color='blue')
plt.title('Cumulative Distribution Function of DebugSize')
plt.xlabel('DebugSize')
plt.ylabel('Cumulative Probability')
plt.grid()

# Highlight a specific threshold (optional)
threshold = 1000  # example threshold
plt.axvline(x=threshold, color='red', linestyle='--', label='Threshold Example')
plt.legend()

plt.show()
```

```python
# Visualize MajorImageVersion counts
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='MajorImageVersion', order=data['MajorImageVersion'].value_counts().index)
plt.title('Counts of MajorImageVersion')
plt.xlabel('MajorImageVersion')
plt.ylabel('Count')
plt.grid()
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Count plot for MajorOSVersion, differentiated by Benign
sns.countplot(data=data, x='MajorOSVersion', hue='Benign', palette='Set2')
plt.title('Distribution of Major OS Versions by Class (Benign vs. Malicious)')
plt.xlabel('Major OS Version')
plt.ylabel('Count')
plt.grid(axis='y')

# Optional: Annotate counts on the bars
for p in plt.gca().patches:
    plt.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='bottom', fontsize=10, color='black', rotation=0)

plt.legend(title='Class', labels=['Benign', 'Malicious'])
plt.xticks(rotation=45)  # Rotate x labels for better readability
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Count plot for NumberOfSections, differentiated by Benign
sns.countplot(data=data, x='NumberOfSections', hue='Benign', palette='Set2')
plt.title('Distribution of Number of Sections by Class (Benign vs. Malicious)')
plt.xlabel('Number of Sections')
plt.ylabel('Count')
plt.grid(axis='y')

# Optional: Annotate counts on the bars
for p in plt.gca().patches:
    plt.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='bottom', fontsize=10, color='black', rotation=0)

plt.legend(title='Class', labels=['Benign', 'Malicious'])
plt.show()
```
